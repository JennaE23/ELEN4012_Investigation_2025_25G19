```python
import sys
sys.path.append('../../')

from investigation_functions import ml_funcs as mlf
from investigation_functions import ml_visualisation_funcs as mlvf
from investigation_functions import data_process_funcs as dpf
from investigation_functions import test_table_funcs as ttf

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.neighbors import KNeighborsClassifier
```

## 4 Qubits

```python
nr_qubits =4

initial_list_4q = ttf.get_HSR_array_all_backends(
    nr_qubits, '../../../',updated_results=True) 

processed_4q = []
for df in initial_list_4q:
    processed_4q.append(mlf.apply_preprosessing(df))
```

### Hardware

```python
k_range = range(3, 26)
k_scores = []

for k in k_range:
    model = KNeighborsClassifier(n_neighbors=k, weights="distance",algorithm='auto', leaf_size=30, p=1)
    fitted_model,score,cv_score = mlf.std_split_fit_and_scores(
        processed_4q[0],model)
    k_scores.append(cv_score.mean())
    print(f"n_neighbors={k}, Accuracy={score}")

plt.plot(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.show()
```
