## 1. Basic Preprocessing

First fetch the labeled data.

```python
dir = "../../"
df_4q_H = dpf.get_expanded_df('Hardware',4,dir)
```

After preprocessing, the data should be ready to be fed directly into a machine learning algorithm. Therefore, redundant features are removed, and all categorical data is label-encoded. Then some feature scaling is performed.

The preprocessing does the following:
1. Drops the experiment_type features because it would not be known in an application setting
2. Drops the 0th column (eg. '0000' for 4 qubits) because it only contains 1s.
3. Transforms all numbers that are the strings datatype to integers (the circuit_type feature mainly)
4. Drops the number of qubits because the data going into the ml algorithms should all have the same number of qubits at a time
5. The totalError feature is divided by the total number of shots (4096) to convert into a relative total error between 0 and 1
6. The circuit_type feature is also normalised
