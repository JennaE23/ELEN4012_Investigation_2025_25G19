```python
def make_file_names(backend_name,is_sim,nr_qubits):
    base_name = nr_qubits +"q_"
    if is_sim:
        base_name = base_name +"fake_"
    base_name = base_name +backend_name

    file_names = []
    for i in range(1,4):
        temp_name = base_name + str(i) +".csv"
        file_names.append(temp_name)
    return file_names
```
