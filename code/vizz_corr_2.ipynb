#### 4q

```python
n_qubits = 4
dir_ = '../'

initial_list_4 = ttf.get_HSR_array_all_backends(n_qubits,dir_,True)
df_H_4 = mlf.apply_preprosessing(initial_list_4[0])
df_S_4 = mlf.apply_preprosessing(initial_list_4[1])
df_R_4 = mlf.apply_preprosessing(initial_list_4[2])
```

```python

HS_4_corr_avg = df_H_4.corrwith(df_S_4,axis = 1).mean()
HR_4_corr_avg=df_H_4.corrwith(df_R_4,axis = 1).mean()
SR_4_corr_avg=df_R_4.corrwith(df_S_4,axis = 1).mean()

print("Avg Correlation H vs S:")
print(HS_4_corr_avg)
print("Avg Correlation H vs R:")
print(HR_4_corr_avg)
print("Avg Correlation R vs S:")
print(SR_4_corr_avg)
```
