```python
    
#this is testing the circuits vs each other for a specific exp_type   
for qubit_nr in [4]:
    initial_list = ttf.get_HSR_array_all_backends(qubit_nr,'../', True) #[df_h,df_s, df_r]
    
    
    for exp_type in initial_list:
        test_table_circuits = ttf.get_circuits_test_table(exp_type)

        for svm_mode in svm_modes:
            ml2csv.run_and_record_test_table_for_mode(test_table_circuits,svm_mode,directory)
    
```

```python
#HSR for one circuit type at a time ('001','010','100')
circuit_options = ['1','2','3']

for qubit_nr in [4]:
  initial_list = ttf.get_HSR_array_all_backends(qubit_nr,'../', True) #[df_h,df_s, df_r]
  
  for circuit_type in circuit_options:
    init_list_circ =[]
    for exp_type in initial_list:
      init_list_circ.append(exp_type[exp_type['circuit_type']==circuit_type])
    test_table_HSR_c = ttf.get_HSR_test_table(init_list_circ)
    for svm_mode in svm_modes:
      ml2csv.run_and_record_test_table_for_mode(test_table_HSR_c,svm_mode,directory)
```

```python
#test backends vs each other (same exp_type)
#(each exp_type has a select mode)

svm_modes_for_exp_type = svm_modes[2:5]
for qubit_nr in [4]:
    initial_list = ttf.get_HSR_array_all_backends(qubit_nr,'../', True) #[df_h,df_s, df_r]
    
   
    for exp_type,mode in zip(initial_list,svm_modes_for_exp_type):
        # self_test = initial_list[initial_list['experiment_type']]
        #split_into_backends
        for backend_combo in backend_combos:
            self_test_b = exp_type[exp_type['backend'].isin(backend_combo)]
            #self_test_b.info()
            ml2csv.run_and_print_ml_results(self_test_b,[self_test_b],mode,directory)
    
```
