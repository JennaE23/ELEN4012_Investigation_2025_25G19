```python
def print_test_table(test_table,Exp_type = True,backends = True, circ_types = True, nans = False):

    for row in test_table:
        print("\nrow")
        for df in row:
            print('\ndf')
            if Exp_type:
                print(df['experiment_type'].unique())
            if backends:
                print(df['backend'].unique())
            if circ_types:
                print(df['circuit_type'].unique())
            if nans:
                print(df.isna().sum())
```
