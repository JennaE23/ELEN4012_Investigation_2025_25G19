- mode 0: good for hardware

- mode 4: good for refreshed 16

```python
from dataclasses import dataclass


@dataclass
class SVM_mode:
    kernel: str = 'linear'
    C: float =0
    coef0: float = 0
    degree: float =3

class SVM_modes:
    mode0 : SVM_mode = SVM_mode()
    mode1: SVM_mode = SVM_mode(kernel = 'poly', C = 450, coef0 = 0.5)
    mode2: SVM_mode = SVM_mode(kernel = 'poly', degree = 5)
    mode3: SVM_mode = SVM_mode(kernel = 'rbf', C = 450)
    mode4: SVM_mode = SVM_mode(kernel = 'poly', C = 450, degree = 5)
        
```

```python
def print_test_table(test_table,Exp_type = True,backends = True, circ_types = True, nans = False):

    for row in test_table:
        print("\nrow")
        for df in row:
            print('\ndf')
            if Exp_type:
                print(df['experiment_type'].unique())
            if backends:
                print(df['backend'].unique())
            if circ_types:
                print(df['circuit_type'].unique())
            if nans:
                print(df.isna().sum())
```
