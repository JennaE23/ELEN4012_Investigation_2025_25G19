```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

from investigation_functions import ml_visualisation_funcs as mvf
```

```python
df_svm_4q = pd.read_csv('../ML_results/SVM/svm_4q.csv')
df_svm_8q = pd.read_csv('../ML_results/SVM/svm_8q.csv')
df_svm_16q_c111 = pd.read_csv('../ML_results/SVM/svm_16q_HSR_c111.csv')

df_knn_4q=pd.read_csv('../ML_results/KNN/knn_4q_v2.csv')
df_knn_8q=pd.read_csv('../ML_results/KNN/knn_8q_v2.csv')
df_knn_16q_c111 = pd.read_csv('../ML_results/KNN/knn_16q_HSR_c111.csv')
```

```python
df_svm_4q = df_svm_4q[df_svm_4q['machines'] == 1111]
df_svm_8q = df_svm_8q[df_svm_8q['machines'] == 1111]

df_knn_4q = df_knn_4q[df_knn_4q['machines'] == 1111]
df_knn_8q = df_knn_8q[df_knn_8q['machines'] == 1111]

df_svm_4q_hsr = mvf.get_df_with_same('tr&v circuits','test circuits',df_svm_4q,False)
df_svm_8q_hsr = mvf.get_df_with_same('tr&v circuits','test circuits',df_svm_8q,False)

df_knn_4q_hsr = mvf.get_df_with_same('tr&v circuits','test circuits',df_knn_4q,False)
df_knn_8q_hsr = mvf.get_df_with_same('tr&v circuits','test circuits',df_knn_8q,False)

df_svm_4q_hsr_c111 = df_svm_4q_hsr[df_svm_4q_hsr['tr&v circuits'] == 111]
df_svm_8q_hsr_c111 = df_svm_8q_hsr[df_svm_8q_hsr['tr&v circuits'] == 111]

df_knn_4q_hsr_c111 = df_knn_4q_hsr[df_knn_4q_hsr['tr&v circuits'] == 111]
df_knn_8q_hsr_c111 = df_knn_8q_hsr[df_knn_8q_hsr['tr&v circuits'] == 111]

df_knn_4q_HSR = pd.concat([df_knn_4q_hsr_c111[df_knn_4q_hsr_c111['tr&v exp_type'] == 'Hardware'],df_knn_4q_hsr_c111[df_knn_4q_hsr_c111['tr&v exp_type'] == 'Simulation'], df_knn_4q_hsr_c111[df_knn_4q_hsr_c111['tr&v exp_type'] == 'Refreshed_Simulation']], ignore_index=True)
df_knn_8q_HSR = pd.concat([df_knn_8q_hsr_c111[df_knn_8q_hsr_c111['tr&v exp_type'] == 'Hardware'],df_knn_8q_hsr_c111[df_knn_8q_hsr_c111['tr&v exp_type'] == 'Simulation'], df_knn_8q_hsr_c111[df_knn_8q_hsr_c111['tr&v exp_type'] == 'Refreshed_Simulation']], ignore_index=True)
df_knn_16q_HSR = pd.concat([df_knn_16q_c111[df_knn_16q_c111['tr&v exp_type'] == 'Hardware'],df_knn_16q_c111[df_knn_16q_c111['tr&v exp_type'] == 'Simulation'], df_knn_16q_c111[df_knn_16q_c111['tr&v exp_type'] == 'Refreshed_Simulation']], ignore_index=True)

df_knn_4q_HSR = pd.concat([df_knn_4q_hsr_c111[df_knn_4q_hsr_c111['test exp_type'] == 'Hardware'],df_knn_4q_hsr_c111[df_knn_4q_hsr_c111['test exp_type'] == 'Simulation'], df_knn_4q_hsr_c111[df_knn_4q_hsr_c111['test exp_type'] == 'Refreshed_Simulation']], ignore_index=True)
df_knn_8q_HSR = pd.concat([df_knn_8q_hsr_c111[df_knn_8q_hsr_c111['test exp_type'] == 'Hardware'],df_knn_8q_hsr_c111[df_knn_8q_hsr_c111['test exp_type'] == 'Simulation'], df_knn_8q_hsr_c111[df_knn_8q_hsr_c111['test exp_type'] == 'Refreshed_Simulation']], ignore_index=True)
df_knn_16q_HSR = pd.concat([df_knn_16q_c111[df_knn_16q_c111['test exp_type'] == 'Hardware'],df_knn_16q_c111[df_knn_16q_c111['test exp_type'] == 'Simulation'], df_knn_16q_c111[df_knn_16q_c111['test exp_type'] == 'Refreshed_Simulation']], ignore_index=True)

df_svm_4q_HSR = pd.concat([df_svm_4q_hsr_c111[df_svm_4q_hsr_c111['tr&v exp_type'] == 'Hardware'],df_svm_4q_hsr_c111[df_svm_4q_hsr_c111['tr&v exp_type'] == 'Simulation'], df_svm_4q_hsr_c111[df_svm_4q_hsr_c111['tr&v exp_type'] == 'Refreshed_Simulation']], ignore_index=True)
df_svm_8q_HSR = pd.concat([df_svm_8q_hsr_c111[df_svm_8q_hsr_c111['tr&v exp_type'] == 'Hardware'],df_svm_8q_hsr_c111[df_svm_8q_hsr_c111['tr&v exp_type'] == 'Simulation'], df_svm_8q_hsr_c111[df_svm_8q_hsr_c111['tr&v exp_type'] == 'Refreshed_Simulation']], ignore_index=True)
df_svm_16q_HSR = pd.concat([df_svm_16q_c111[df_svm_16q_c111['tr&v exp_type'] == 'Hardware'],df_svm_16q_c111[df_svm_16q_c111['tr&v exp_type'] == 'Simulation'], df_svm_16q_c111[df_svm_16q_c111['tr&v exp_type'] == 'Refreshed_Simulation']], ignore_index=True)

df_svm_4q_HSR = pd.concat([df_svm_4q_hsr_c111[df_svm_4q_hsr_c111['test exp_type'] == 'Hardware'],df_svm_4q_hsr_c111[df_svm_4q_hsr_c111['test exp_type'] == 'Simulation'], df_svm_4q_hsr_c111[df_svm_4q_hsr_c111['test exp_type'] == 'Refreshed_Simulation']], ignore_index=True)
df_svm_8q_HSR = pd.concat([df_svm_8q_hsr_c111[df_svm_8q_hsr_c111['test exp_type'] == 'Hardware'],df_svm_8q_hsr_c111[df_svm_8q_hsr_c111['test exp_type'] == 'Simulation'], df_svm_8q_hsr_c111[df_svm_8q_hsr_c111['test exp_type'] == 'Refreshed_Simulation']], ignore_index=True)
df_svm_16q_HSR = pd.concat([df_svm_16q_c111[df_svm_16q_c111['test exp_type'] == 'Hardware'],df_svm_16q_c111[df_svm_16q_c111['test exp_type'] == 'Simulation'], df_svm_16q_c111[df_svm_16q_c111['test exp_type'] == 'Refreshed_Simulation']], ignore_index=True)
```

```python
df_knn_full = pd.concat([df_knn_4q_HSR,df_knn_8q_HSR,df_knn_16q_HSR], ignore_index=True)
df_svm_full = pd.concat([df_svm_4q_HSR,df_svm_8q_HSR,df_svm_16q_HSR], ignore_index=True)
```

```python
df_svm_full = pd.concat([df_svm_full[df_svm_full['param settings'] == 4]], ignore_index=True)
df_svm_full = df_svm_full[df_svm_full['tr&v exp_type'] != "Simulation Refreshed_Simulation"]
df_svm_full = df_svm_full[df_svm_full['tr&v exp_type'] != "Refreshed_Simulation Hardware"]
df_svm_full = df_svm_full[df_svm_full['tr&v exp_type'] != "Hardware Simulation"]
df_svm_full = df_svm_full[df_svm_full['tr&v exp_type'] != "Sim and Refreshed"]
df_svm_full.head(30)
```

```python
df_knn_full = pd.concat([df_knn_full[df_knn_full['param settings'] == 4]], ignore_index=True)
df_knn_full = df_knn_full[df_knn_full['tr&v exp_type'] != "Simulation Refreshed_Simulation"]
df_knn_full = df_knn_full[df_knn_full['tr&v exp_type'] != "Refreshed_Simulation Hardware"]
df_knn_full = df_knn_full[df_knn_full['tr&v exp_type'] != "Hardware Simulation"]
df_knn_full = df_knn_full[df_knn_full['tr&v exp_type'] != "Sim and Refreshed"]
df_knn_full.head(30)
```

```python
df_knn_full['accuracy'] = df_knn_full['accuracy'] * 100
df_svm_full['accuracy'] = df_svm_full['accuracy'] * 100
```

```python
fontsize_ = 30

plt.figure(figsize=(10,6))
sns.barplot(
    data = df_knn_full[df_knn_full['nr_qubits'] == 4],
    x = 'tr&v exp_type',
    y = 'accuracy',
    hue= 'test exp_type'    
)
plt.ylim(tuple([0,100]))
# plt.title('ML Model Accuracy vs Size of Algorithm')
plt.xlabel('Train Experiment Type', fontsize = fontsize_)
plt.xticks([0,1,2],["H", "S", "SR"],fontsize = fontsize_)
plt.yticks(fontsize = fontsize_)
plt.ylabel('Accuracy (%)', fontsize = fontsize_)
plt.legend(title='Test Experiment Type',loc='center left', bbox_to_anchor=(1, 0.5), fontsize = fontsize_, title_fontsize = fontsize_)
plt.show()
```
