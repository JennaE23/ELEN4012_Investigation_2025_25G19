```python
def get_meta_dataframe(type_data, nr_qubits):
    big_df =  meta_dataframe_functions.blank_meta_df()
    meta_dataframe_functions.load_meta_df(big_df,type_data)

    df_qubits = big_df[big_df['nr_qubits']==nr_qubits]

    # for nr_qubits in qubit_types:
    #     df_arr.append(big_df[big_df['nr_qubits']==nr_qubits])

    for i in range((df_qubits.size(0))): #for each row
        df = data_process_funcs.create_processed_df(df_qubits['file_path'][i])
        labels = np.empty(df.size(0))
        labels.fill([df_qubits['circuit_type','backend','nr_qubits'][i]])
        
        df.insert(loc=0, column = ['circuit_type','backend','nr_qubits'], value =labels)

    return df
```
