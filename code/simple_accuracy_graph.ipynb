```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

from investigation_functions import ml_visualisation_funcs as mvf
```

```python
df_svm_4q = pd.read_csv('../ML_results/SVM/svm_4q.csv')
df_svm_8q = pd.read_csv('../ML_results/SVM/svm_8q.csv')
df_svm_16q_c111 = pd.read_csv('../ML_results/SVM/svm_16q_HSR_c111.csv')
```

```python
df_knn_4q = pd.read_csv('../ML_results/KNN/knn_4q.csv')
df_knn_8q = pd.read_csv('../ML_results/KNN/knn_8q.csv')
df_knn_16q_c111 = pd.read_csv('../ML_results/KNN/knn_16q_HSR_c111.csv')
```

```python
df_svm_4q = df_svm_4q[df_svm_4q['machines'] == 1111]
df_svm_8q = df_svm_8q[df_svm_8q['machines'] == 1111]
```

```python
df_knn_4q = df_knn_4q[df_knn_4q['machines'] == 1111]
df_knn_8q = df_knn_8q[df_knn_8q['machines'] == 1111]
```

```python
df_svm_4q_hsr_c111 = df_svm_4q[df_svm_4q['tr&v circuits'] == 111]
df_svm_8q_hsr_c111 = df_svm_8q[df_svm_8q['tr&v circuits'] == 111]
```

```python
df_knn_4q_hsr_c111 = df_knn_4q[df_knn_4q['tr&v circuits'] == 111]
df_knn_8q_hsr_c111 = df_knn_8q[df_knn_8q['tr&v circuits'] == 111]
```

```python
df_knn_4q_hsr_c111.info()
```

```python
df_4q = pd.concat([df_svm_4q_hsr_c111.drop('kernal', axis=1),df_knn_4q_hsr_c111.drop('n_neighbors', axis=1)],ignore_index=True)
```

```python
df_4q.info()
```

```python
df_8q = pd.concat([df_svm_8q_hsr_c111.drop('kernal', axis = 1),df_knn_8q_hsr_c111.drop('n_neighbors',axis =1)],ignore_index=True)
# df_16q = pd.concat([df_svm_16q_c111,df_knn_16q_c111],ignore_index=True)
df_8q.info()
```

```python
df_16q = pd.concat([df_svm_16q_c111.drop('kernal', axis=1),df_knn_16q_c111.drop('n_neighbors', axis=1)],ignore_index=True)
df_16q.info()
```

```python
df_16q.head()
```

```python
df_4q = df_4q[df_4q['tr&v exp_type'] == df_4q['test exp_type']]
df_4q.info()
```

```python
df_4q.head()
```

```python
df_8q = df_8q[df_8q['tr&v exp_type'] == df_8q['test exp_type']]
df_16q = df_16q[df_16q['tr&v exp_type'] == df_16q['test exp_type']]
```

```python
df_8q.info()
```

```python
df_full = pd.concat([df_4q,df_8q,df_16q],ignore_index=True)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(
    data = df_full,
    x = 'nr_qubits',
    y = 'accuracy',
    hue= 'tr&v exp_type'    
)
plt.title('ML Model Accuracy vs Number of Qubits')
plt.xlabel('Number of Qubits')
plt.ylabel('Accuracy')
plt.legend(title='Experiment Type',loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
