```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
data1 = {'KNN': 100, 'SVM': 97}
df1 = pd.DataFrame(list(data1.items()), columns=['Algorithm', 'Accuracy'])
df1.head()
```

```python
data2 = {'KNN': 100, 'SVM': 67}
df2 = pd.DataFrame(list(data1.items()), columns=['Algorithm', 'Accuracy'])

data3 = {'KNN': 38, 'SVM': 33}
df3 = pd.DataFrame(list(data1.items()), columns=['Algorithm', 'Accuracy'])
```

```python
fontsize_ = 30

plt.figure(figsize=(10,6))
sns.barplot(
    data=data1,
    # x='Algorithm',
    # y='Accuracy',
)
plt.xlabel('Algorithm', fontsize = fontsize_)
plt.ylabel('Accuracy (%)', fontsize = fontsize_)
plt.ylim(tuple([0,100]))
plt.xticks(fontsize = fontsize_)
plt.yticks(fontsize = fontsize_)
plt.title('Trained on C1 Tested on C2', fontsize = fontsize_)
plt.show()
```

```python
fontsize_ = 30

plt.figure(figsize=(10,6))
sns.barplot(
    data=data2,
    # x='Algorithm',
    # y='Accuracy',
)
plt.xlabel('Algorithm', fontsize = fontsize_)
plt.ylabel('Accuracy (%)', fontsize = fontsize_)
plt.ylim(tuple([0,100]))
plt.xticks(fontsize = fontsize_)
plt.yticks(fontsize = fontsize_)
plt.title('Trained on C2 Tested on C1', fontsize = fontsize_)
plt.show()
```
